```python
import numpy as np
import tkinter as tk

win = tk.Tk()
win.geometry("700x500")
win.title("Functional Dependency Calculator")
attr_label = tk.Label(win, text="Relation R(", font=("default", 12))
attr_label.grid(row=0, column=0)
attr_label_close = tk.Label(win, text=")", font=("default", 12))
attr_label_close.grid(row=0, column=2)
attr = tk.StringVar()
attr_entry = tk.Entry(win, textvariable=attr, font=("default", 12))
attr_entry.grid(row=0, column=1)
attr_entry.focus()

def fd_calculate():
    a = len(attr.get())  #No of attributes
    x = []
    y = []
    xy = fd_text.get(1.0, "end-1c").replace(' ', '').splitlines()
    n = len(xy)     #No of FDs
    attr_str = attr_entry.get().replace(' ', '')
    attr_list = list(attr_str)
    for line in xy:
        temp = line.split("->")
        x.append(temp[0])
        y.append(temp[1])
    xside = np.chararray((n, a), unicode=True)
    yside = np.chararray((n, a), unicode=True)
    for i in range(n):
        xatt = list(x[i])
        yatt = list(y[i])
        lx = len(xatt)
        for j in range(lx):
            xside[i][j] = xatt[j]
        ly = len(yatt)
        for j in range(ly):
            yside[i][j] = yatt[j]

    closure = []

    for i in range(a):
        s1 = attr_list[i]
        s2 = s1
        for m in range(n):
            for j in range(i + m, i + m + n):
                flag = 1
                for k in range(len(xside[j % n])):
                    if (s1.find(xside[j % n][k]) == -1):
                        flag = 0
                        break
                if flag == 1:
                    for k in range(len(yside[j % n])):
                        if (s1.find(yside[j % n][k]) == -1):
                            s1 = s1 + yside[j % n][k]

        closure.append(s1)

        closure_label['text'] += "\nClosure of " + s2 + " is: " + s1


    ck = []

    for i in range(a):
        s3 = ""
        flag1 = 0
        ck1 = attr_list[i]
        for j in range(i, i + a):

            s4=s3
            s3 = s3 + closure[j % a]

            if len(set(s4))<len(set(s3)):
                if len(set(s3)) < a:
                    ck1 = ck1 + attr_list[j % a]

                if len(set(s3)) == a:
                    flag1 = 1
                    ck1 = ck1 + attr_list[j % a]
                    ck2 = set(ck1)
                    break

        if flag1 == 1:
            ck.append(ck2)
        #print(ck)

    unique_ck = np.chararray((len(closure), a), unicode=True)

    for i in range(len(ck)):
        j = 0
        for x in ck[i]:
            if x not in unique_ck[i]:
                unique_ck[i][j] = x
                j = j + 1

    #print(unique_ck)


    mini = a

    for i in range(len(ck)):
        c = 0
        for j in range(a):
            if unique_ck[i][j] != '':
                c = c + 1
        if c < mini:
            mini = c
            pos = i

    for j in range(a):
        if j == 0:
            ck_label['text'] += "\n" + unique_ck[pos][j]
        else:
            ck_label['text'] += unique_ck[pos][j]
            
    candidate_key=[]

    for i in range(len(ck)):
        m = 0
        c_k = ""
        for j in range(a):
            if unique_ck[i][j] != '':
                m = m + 1
        
        if m==mini and pos!=i:
            for j in range(m):
                #print(unique_ck[i][j],end="") 
                c_k = c_k + unique_ck[i][j]

            if c_k not in candidate_key:
                candidate_key.append(c_k)
                ck_label['text'] += "\n" + c_k


fd_label = tk.Label(win, text="Enter Function Dependencies: ", font=("default", 12))
fd_label.grid(row=1, column=0)
fd_text = tk.Text(win, height=10, width=10, font=("default", 12))
fd_text.grid(row=1, column=1)

fd_button = tk.Button(win, text="Calculate", command=fd_calculate, font=("default", 12))
fd_button.grid(row=1, column=2)

closure_label = tk.Label(win, text="Attribute Closures", font=("default", 12))
closure_label.grid(row=3, column=0, sticky='N')
ck_label = tk.Label(win, text="Candidate Key", font=("default", 12))
ck_label.grid(row=3, column=1, sticky='N')

win.mainloop()
```
